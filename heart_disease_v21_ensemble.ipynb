# V21: Full Blending with Previous Submission Files
Pure CD Blend + CD95_AB05 Variant

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')
sub_d = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/D.csv')

def rank_norm(preds):
    return pd.Series(preds).rank(pct=True).values

c_rn = rank_norm(sub_c[TARGET].values)
d_rn = rank_norm(sub_d[TARGET].values)

# Option 1: Pure CD
pure_cd = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(0.5 * c_rn + 0.5 * d_rn, 0, 1)
})
pure_cd.to_csv('/kaggle/working/pure_cd.csv', index=False)

# Option 2: CD95 + AB05
sub_a = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/A.csv')
sub_b = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/B.csv')
a_rn = rank_norm(sub_a[TARGET].values)
b_rn = rank_norm(sub_b[TARGET].values)
ab = 0.5 * a_rn + 0.5 * b_rn
cd = 0.5 * c_rn + 0.5 * d_rn

cd95_ab05 = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(0.95 * cd + 0.05 * ab, 0, 1)
})
cd95_ab05.to_csv('/kaggle/working/cd95_ab05.csv', index=False)

print("âœ… Both files saved!")
print(f"Pure CD range: [{pure_cd['Heart Disease'].min():.4f}, {pure_cd['Heart Disease'].max():.4f}]")
```
