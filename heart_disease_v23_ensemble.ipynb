# V23: Raw C vs D Blends - No Rank Normalization
Testing raw values: C, D, C70_D30, D70_C30, C90_D10, D90_C10

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')
sub_d = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/D.csv')

c = sub_c[TARGET].values
d = sub_d[TARGET].values

# All 6 options — NO rank normalization
files = {
    'raw_c':     c,
    'raw_d':     d,
    'c70_d30':   0.7 * c + 0.3 * d,
    'd70_c30':   0.7 * d + 0.3 * c,
    'c90_d10':   0.9 * c + 0.1 * d,
    'd90_c10':   0.9 * d + 0.1 * c,
}

for name, preds in files.items():
    sub = pd.DataFrame({
        'id': sub_c['id'],
        'Heart Disease': np.clip(preds, 0, 1)
    })
    sub.to_csv(f'/kaggle/working/{name}.csv', index=False)
    print(f"✅ Saved {name}.csv")

print("\nDone! Submit in this order:")
print("1. raw_c.csv")
print("2. raw_d.csv")
print("3. c70_d30.csv")
print("4. d70_c30.csv")
print("5. c90_d10.csv")
print("6. d90_c10.csv")
```
