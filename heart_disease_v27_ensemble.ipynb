# V27: E + F Blending
F pure submission + E/F blend. Main: F (pure)

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

# Load files
sub_e = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/E.csv')
sub_f = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/F.csv')

print(f"E range: [{sub_e[TARGET].min():.4f}, {sub_e[TARGET].max():.4f}]")
print(f"F range: [{sub_f[TARGET].min():.4f}, {sub_f[TARGET].max():.4f}]")

e = sub_e[TARGET].values
f = sub_f[TARGET].values

pure_f = pd.DataFrame({
    'id': sub_f['id'],
    'Heart Disease': np.clip(f, 0, 1)
})
pure_f.to_csv('/kaggle/working/submission.csv', index=False)
print("✅ Saved submission.csv")

ef_blend = pd.DataFrame({
    'id': sub_f['id'],
    'Heart Disease': np.clip(0.5 * e + 0.5 * f, 0, 1)
})
ef_blend.to_csv('/kaggle/working/ef_blend.csv', index=False)
print("✅ Saved ef_blend.csv")

print(pure_f.head(5))
```
