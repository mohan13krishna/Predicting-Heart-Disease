# V22: C vs D Variants - No Rank Normalization
Testing C70_D30, D70_C30, Only C, Only D

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')
sub_d = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/D.csv')

def rank_norm(preds):
    return pd.Series(preds).rank(pct=True).values

# Option 1: Only C
only_c = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(rank_norm(sub_c[TARGET].values), 0, 1)
})
only_c.to_csv('/kaggle/working/only_c.csv', index=False)

# Option 2: Only D
only_d = pd.DataFrame({
    'id': sub_d['id'],
    'Heart Disease': np.clip(rank_norm(sub_d[TARGET].values), 0, 1)
})
only_d.to_csv('/kaggle/working/only_d.csv', index=False)

# Option 3: C70 + D30
c70_d30 = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(0.7 * rank_norm(sub_c[TARGET].values) + 0.3 * rank_norm(sub_d[TARGET].values), 0, 1)
})
c70_d30.to_csv('/kaggle/working/c70_d30.csv', index=False)

print("âœ… All files saved!")
```
