# V24: C with Micro-Adjustments from B
C99_B01 and C98_B02 variants

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')
sub_b = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/B.csv')

c = sub_c[TARGET].values
b = sub_b[TARGET].values

# Try C with very small amount of our best model
c99_b01 = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(0.99 * c + 0.01 * b, 0, 1)
})
c99_b01.to_csv('/kaggle/working/c99_b01.csv', index=False)

c98_b02 = pd.DataFrame({
    'id': sub_c['id'],
    'Heart Disease': np.clip(0.98 * c + 0.02 * b, 0, 1)
})
c98_b02.to_csv('/kaggle/working/c98_b02.csv', index=False)

print("âœ… Done!")
```
