# V28: F with Micro-Perturbations (E + C)
7 micro-blend variants with F as base. Main: F99_E01

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

# Load files
sub_f = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/F.csv')
sub_e = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/E.csv')
sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')

print(f"F range: [{sub_f[TARGET].min():.4f}, {sub_f[TARGET].max():.4f}]")
print(f"E range: [{sub_e[TARGET].min():.4f}, {sub_e[TARGET].max():.4f}]")
print(f"C range: [{sub_c[TARGET].min():.4f}, {sub_c[TARGET].max():.4f}]")

f = sub_f[TARGET].values
e = sub_e[TARGET].values
c = sub_c[TARGET].values

#  Micro-perturbation blends
blends = {
    'f99_e01':      0.99 * f + 0.01 * e,
    'f99_c01':      0.99 * f + 0.01 * c,
    'f98_e01_c01':  0.98 * f + 0.01 * e + 0.01 * c,
    'f97_e02_c01':  0.97 * f + 0.02 * e + 0.01 * c,
    'f97_e01_c02':  0.97 * f + 0.01 * e + 0.02 * c,
    'f95_e05':      0.95 * f + 0.05 * e,
    'f95_c05':      0.95 * f + 0.05 * c,
}

for name, preds in blends.items():
    sub = pd.DataFrame({
        'id': sub_f['id'],
        'Heart Disease': np.clip(preds, 0, 1)
    })
    sub.to_csv(f'/kaggle/working/{name}.csv', index=False)
    print(f"âœ… Saved {name}.csv")

final = pd.DataFrame({
    'id': sub_f['id'],
    'Heart Disease': np.clip(0.99 * f + 0.01 * e, 0, 1)
})
final.to_csv('/kaggle/working/submission.csv', index=False)

print(final.head(5))
print(f"\nRange: [{final['Heart Disease'].min():.6f}, {final['Heart Disease'].max():.6f}]")
```
