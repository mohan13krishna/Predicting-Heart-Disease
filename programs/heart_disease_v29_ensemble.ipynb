# V29: F90_C10 Blend
Simple F90 + C10 blend

```python
import pandas as pd
import numpy as np

TARGET = 'Heart Disease'

# Load files
sub_f = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/F.csv')
sub_c = pd.read_csv('/kaggle/input/datasets/mohankrishnathalla/predicting-heart-disease-submission/C.csv')

f = sub_f[TARGET].values
c = sub_c[TARGET].values

final = pd.DataFrame({
    'id': sub_f['id'],
    'Heart Disease': np.clip(0.90 * f + 0.10 * c, 0, 1)
})
final.to_csv('/kaggle/working/submission.csv', index=False)
print("âœ… Saved!")
```
